```python
import tkinter as tk
from tkinter import ttk
import pickle
import pandas as pd

# Load model
model = pickle.load(open("model.pkl", "rb"))

def predict():
    try:
        # Convert inputs
        gender_val = 1 if gender.get() == "Male" else 0
        
        geo_germany = 1 if geography.get() == "Germany" else 0
        geo_spain = 1 if geography.get() == "Spain" else 0

        data = [[
            int(cs.get()),
            gender_val,
            int(age.get()),
            int(tenure.get()),
            float(balance.get()),
            int(products.get()),
            int(card.get()),
            int(active.get()),
            float(salary.get()),
            geo_germany,
            geo_spain
        ]]

        df = pd.DataFrame(data, columns=[
            'CreditScore','Gender','Age','Tenure','Balance',
            'NumOfProducts','HasCrCard','IsActiveMember',
            'EstimatedSalary','Geography_Germany','Geography_Spain'
        ])

        result = model.predict(df)[0]

        if result == 1:
            output.config(text="Customer will EXIT ", fg="red")
        else:
            output.config(text="Customer will STAY ", fg="green")

    except:
        output.config(text="Please enter valid data", fg="orange")


# GUI window
root = tk.Tk()
root.title("Customer Churn Prediction")
root.geometry("400x500")

# Labels + Inputs
def create_input(label_text):
    tk.Label(root, text=label_text).pack()
    entry = tk.Entry(root)
    entry.pack()
    return entry

cs = create_input("Credit Score")
age = create_input("Age")
tenure = create_input("Tenure")
balance = create_input("Balance")
products = create_input("Number of Products")
salary = create_input("Estimated Salary")

# Dropdowns
tk.Label(root, text="Gender").pack()
gender = ttk.Combobox(root, values=["Male", "Female"])
gender.pack()

tk.Label(root, text="Geography").pack()
geography = ttk.Combobox(root, values=["France", "Germany", "Spain"])
geography.pack()

tk.Label(root, text="Has Credit Card (1=Yes, 0=No)").pack()
card = create_input("")

tk.Label(root, text="Is Active Member (1=Yes, 0=No)").pack()
active = create_input("")

# Button
tk.Button(root, text="Predict", command=predict).pack(pady=10)

# Output
output = tk.Label(root, text="", font=("Arial", 12))
output.pack()

root.mainloop()
```
